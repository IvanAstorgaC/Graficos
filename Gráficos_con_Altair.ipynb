```python
alt.Chart(chile).mark_bar(color='#e6a6be').encode(
    x=alt.X('year:O',title='',axis=alt.Axis(labelAngle=45,labelFontSize=12)),
    y=alt.Y('pop:Q',title='Poblacin', axis=alt.Axis(format='s',labelFontSize=12)),
    tooltip=alt.Tooltip('pop')
).properties(
    title=f'Poblacin de Chile',
    width=600,
)
#si queremos hacerlo interactivo usamos tooltip=alt.Tooltip
```

```python
#grafico de barras y texto
#primero las barras
barras=alt.Chart(chile).mark_bar(color='#e6a6be').encode(
    x=alt.X('year:O',title='',axis=alt.Axis(labelAngle=45,labelFontSize=12)),
    y=alt.Y('pop:Q',title='Poblacin', axis=alt.Axis(format='s',labelFontSize=12,labels=False,grid=False, domain=False)),#labels quita las etiquetas, grid quita la malla y domain quitaa la barra
    tooltip=alt.Tooltip('pop')
)

#ahora creamos el texto

texto=barras.mark_text(dy= -5).encode(
    text=alt.Text('pop:Q',format='.2s')
)

#los atributos se dan despues de los graficos
(barras+texto).properties(
    title=f'Poblacin de Chile (1955-2005)',
    width=600,
)
```

```python
#grafico de barras y texto
#primero las barras
barras_2=alt.Chart(mexico).mark_bar(color='#e6a6be').encode(
    x=alt.X('year:O',title='',axis=alt.Axis(labelAngle=45,labelFontSize=12)),
    y=alt.Y('pop:Q',title='Poblacin', axis=alt.Axis(format='s',labelFontSize=12,labels=False,grid=False, domain=False)),#labels quita las etiquetas, grid quita la malla y domain quitaa la barra
    tooltip=alt.Tooltip('pop')
)

#ahora creamos el texto

texto_2=barras_2.mark_text(dy= -5).encode(
    text=alt.Text('pop:Q',format='.2s')
)

#los atributos se dan despues de los graficos
(barras_2+texto_2).properties(
    title=f'Poblacin de Mxico (1955-2005)',
    width=600,
)
```

```python
from vega_datasets import data

source = data.cars()

brush = alt.selection(type='interval') # selector

points = alt.Chart(source).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    color=alt.condition(brush, 'Origin:N', alt.value('lightgray'))
).add_selection(
    brush
) # grafico de puntos

bars = alt.Chart(source).mark_bar().encode(
    y='Origin:N',
    color='Origin:N',
    x='mean(Miles_per_Gallon):Q'
).transform_filter(
    brush
)# esto ultimo sirve para que solo use los datos en la seleccion del brush

points & bars
```

```python
# Regresin

#Creamos la base
base = alt.Chart(source).mark_circle().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
).properties(
    width=300,
    height=300
)

#Regresin Lineal

polynomial_fit_grade_1 = [
    base.transform_regression(
        "Horsepower", "Miles_per_Gallon", method="poly", order=1, as_=["Horsepower", 'Regresin Lineal']
    )
    .mark_line()
    .transform_fold(['Regresin Lineal'], as_=["degree", "Miles_per_Gallon"])
    .encode(alt.Color("degree:N",legend=alt.Legend(title='')))
]

reg1=alt.layer(base, *polynomial_fit_grade_1)
reg1
```

```python
# Regresin polinmica

#Grado 3

polynomial_fit_grade_3 = [
    base.transform_regression(
        "Horsepower", "Miles_per_Gallon", method="poly", order=3, as_=["Horsepower", 'Polinomial Orden 3']
    )
    .mark_line()
    .transform_fold(['Polinomial Orden 3'], as_=["degree", "Miles_per_Gallon"])
    .encode(alt.Color("degree:N",legend=alt.Legend(title='')))
]

reg2=alt.layer(base, *polynomial_fit_grade_3)
reg2
```
